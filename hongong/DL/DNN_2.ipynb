##  1️⃣ 과대적합을 막기 위한 수단
- Optimizer - Adam
- Dropout layer
- EarlyStopping (Callback)

Data Preparation

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

train_scaled = X_train/255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, y_train, test_size=0.2, random_state=42
)
```

신경망 모델 정의

```python
def model_fn(a_layer=None):
  model = keras.Sequential()
  model.add(keras.layers.Flatten(input_shape=(28, 28)))
  model.add(keras.layers.Dense(100, activation='relu', name='hidden'))
  if a_layer:
    model.add(a_layer)
  model.add(keras.layers.Dense(10, activation='softmax', name='output'))
  return model
```

```python
model = model_fn()
model.summary()
```

fit()의 History 객체 return

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')

history = model.fit(train_scaled, train_target, epochs=5, verbose=0)
```

```python
history.history.keys()
```

Loss Plot

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

Accuracy Plot

```python
plt.plot(history.history['accuracy'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

epoch가 진행될수록 loss는 줄고, accuracy는 높아짐을 알 수 있다.

epoch를 더 늘려 진행하고 loss와 accuracy를 시각화해보자.

```python
# epoch=20

model = model_fn()
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0)

plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Loss Plot')
plt.show()
```

epoch가 진행될수록 loss가 줄어든다. 하지만, 과대적합될 가능성이 있으므로 train뿐 아니라 validation set을 통해 검증해야한다.

```python
model = model_fn()

model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0,
          validation_data=(val_scaled, val_target))

history.history.keys()
```

시각화 하여 과대적합, 과소적합 여부를 확인하자.

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train', 'validation'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

validation set의 loss가 증가하는 시점이 과대적합되었다 볼 수 있다. 이 시점을 늦추면, train set과 validationset에서의 성능이 좋아진다.

과대적합을 완화하기 위해
- Optimizer - Adam
- 규제 - Dropout
- 조기 종료 (callback)

방법을 적용해볼 수 있다.

### [1]  **Adam** Optimizer

default Optimizer : RMSprop

```python
model = model_fn()

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=20, verbose=0,
          validation_data=(val_scaled, val_target))

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'validation'])
plt.show()
```

Adam optimizer를 사용함으로써 이전에 비해 validation loss가 대체적으로 감소 그래프를 그리므로 과대적합이 훨씬 줄어들었다 할 수 있다.

### [2] 신경망 과대적합 규제 - **Drop out**

```python
model = model_fn(keras.layers.Dropout(0.3))

model.summary()
```

Train 및 검증

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy',)
history = model.fit(train_scaled, train_target, epochs=20, verbose=0,
                    validation_data=(val_scaled, val_target))

print(history.history.keys())
```

Plot

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.legend(['validation', 'train'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

앞의 모델에 보다 훨씬 과대적합을 줄인 것을 확인할 수 있고, epoch=10이 적절한 epoch인 것 같다.

##  2️⃣ 모델 저장, 복원
epoch=10인 모델을 다시 정의하고 학습시켜보자

```python
model = model_fn(keras.layers.Dropout(0.3))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
history = model.fit(train_scaled, train_target, epochs=10, verbose=0,
                    validation_data=(val_scaled, val_target))

print(history.history['val_accuracy'])
```

### [2] 모델 복원

2-1. 모델 파라미터

정의한 모델 구조가 저장한 모델 구조와 일치해야 복원 가능하다.

```python
model = model_fn(keras.layers.Dropout(0.3))

model.load_weights('model-weights.keras')
```

정확도 비교

2-2. 모델 전체

##  3️⃣ Callback

훈련 중간에 작업이 가능하다.
- ModelCheckpoint
- EarlyStopping

3-1. **ModelCheckpoint**

매 epoch마다 모델을 **저장**한다.

```python
model = model_fn(keras.layers.Dropout(0.3))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
# callback 객체
checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.keras', save_best_only=True)
model.fit(train_scaled, train_target, epochs=20, verbose=0,
          validation_data=(val_scaled, val_target), callbacks=[checkpoint_cb])
```

```python
# 저장된 best-model 객체 성능 평가
model = keras.models.load_model('best-model.keras')
model.evaluate(val_scaled, val_target)
```

3-2. **EarlyStopping**

과대적합이 되기 **전** 훈련을 **종료**한다.

- ModelCheckpoint 콜백 & EarlyStopping 콜백
  - best 모델 저장
  - 과대적합되기 전 훈련 중단 -> 리소스 절약
  - best 모델일 때의 가중치 확인 가능
두 callback을 함께 쓰면 다음의 편리함이 있다.👍

```python
model = model_fn(keras.layers.Dropout(0.3))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)

history = model.fit(train_scaled, train_target, epochs=20, verbose=0,
          validation_data=(val_scaled, val_target),
          callbacks=[checkpoint_cb, early_stopping_cb])
```

```python
# 저장된 best model 성능 평가
model = keras.models.load_model('best-model.keras')
model.evaluate(val_scaled, val_target)
```

```python
# 진행된 epoch 수
print(early_stopping_cb.stopped_epoch)
```

plot

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['validation', 'train'])
plt.show()
```

EarlyStopping callback을 사용하면 epoch 수를 안심하고 크게 설정해도 된다는 장점이 있다. 과대적합되기 이전에 훈련이 종료되기 때문이다.
