## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

- feature **matrix**
- dependent **vector**

```
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ai-study/Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```
print(X)
```

```
print(y)
```

## Taking care of missing data

- ignore null data
- replace null with avg value


```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:3])
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```
print(X)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# transformation type : encoder | encoder type | column index
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```
print(X)
```

### Encoding the Dependent Variable
- labelEncoding (binary)

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

>📌 Feature scaling은 **train, test split 이후** 이뤄져야 한다.<br/>
>그 이유는, test data는 train data에 포함되지 않아, 모델 성능을 평가하는 데 쓰여야 하기 때문이다.<br/>
>split 이전에 하게 되면, train data에 test data의 특성이 반영되게 되므로, 그 의미가 희석된다.

> 📌 모든 머신러닝 알고리즘에 필요한 작업은 아니다.

> 📌 Encoding된 feature에는 적용해선 안된다. <br/>원래 카테고리 의미를 잃게 된다. 오직 **numerical data**에만 적용해야 한다.




```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```
print(X_train)
```

```
print(X_test)
```
